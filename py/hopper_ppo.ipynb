```python
from pathlib import Path

from rl.hopper.config import PPOConfig
from rl.hopper.plotting import plot_training_history
from rl.hopper.training import train
```

```python
config = PPOConfig(
    env_id="Hopper-v5",
    num_envs=8,
    total_timesteps=1_000_000,
    steps_per_rollout=2048,
    batch_size=64,
    n_epochs=10,
    lr=3e-4,
    gamma=0.99,
    gae_lambda=0.95,
    clip_eps=0.2,
    ent_coef=0.0,
    vf_coef=0.5,
    target_kl=0.015,
    seed=42,
    output_dir=Path("artifacts/hopper_ppo"),
)
config
```

```python
result = train(config)
result.checkpoint_path
```

```python
plot_training_history(result)
```

```python
from pathlib import Path
from IPython.display import display
from rl.hopper.export_mjroll import export_rollout
from rl.hopper.visualization import HopperVisualization

rollout_path = Path("artifacts/hopper_ppo/hopper-rollout.mjroll")
checkpoint_path = result.best_checkpoint_path or result.checkpoint_path

export_rollout(
    env_id=config.env_id,
    checkpoint=checkpoint_path,
    out_path=rollout_path,
    episodes=1,
    deterministic=True,
    policy_hidden_dims=config.policy_hidden_dims,
    value_hidden_dims=config.value_hidden_dims,
    activation=config.activation,
    log_std_init=config.log_std_init,
    ortho_init=config.ortho_init,
    seed=config.seed,
)

display(HopperVisualization.from_rollout_file(rollout_path))
```
